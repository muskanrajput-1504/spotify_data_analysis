```python
import pandas as pd
import zipfile

zip_file_path = '/content/archive (1).zip'
with zipfile.ZipFile(zip_file_path, 'r') as z:
    z.extractall('.')

df = pd.read_csv('playlist_2010to2023.csv', encoding='latin1')
df.head()
```

```python
import matplotlib.pyplot as plt
top_artists = df['artist_name'].value_counts().head(10)
plt.figure(figsize=(10,6))
top_artists.plot(kind='bar', color='mediumpurple', edgecolor='black')
plt.title('Top 10 Artists with the most songs in the Datasets', fontsize=14, fontweight='bold')
plt.xlabel('Artist name', fontsize=12)
plt.ylabel('Number of songs', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
